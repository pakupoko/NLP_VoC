```python
def label_feature(data):
    data['feature'] = 0
    f_list = ['pixel','lens','optical','memory','burst','battery','focus','LCD','compression','flash']
    for idx in range(len(data)):
        features = []
        for f in f_list:
            if f in data.loc[idx, 'review']:
                features.append(f)
        
        f_str = ','.join(features)
        data.loc[idx, 'feature'] = f_str
    
    return data
```
