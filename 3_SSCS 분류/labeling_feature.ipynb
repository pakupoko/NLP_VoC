```python
#apply 활용
def label_feature(data):
    
    data['feature'] = data['review']
    
    def find_feature(text):
        fls = []
        f_list = ['pixel','lens','optical','memory','burst','battery','focus','lcd','compression','flash']
        for f in f_list:
            if f in text:
                fls.append(f)
        if len(fls) == 0 :
            return ''
        else:
            return ','.join(fls)
        
    data['feature']=data['feature'].apply(find_feature)
    
    return data
```
